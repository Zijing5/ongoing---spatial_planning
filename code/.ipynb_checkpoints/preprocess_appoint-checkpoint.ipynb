```python
import pandas as pd
import numpy as np
```

```python
appoint = pd.read_excel('../file/使用情况测试.xlsx')
```

```python
appoint
```

```python
# 删除前两行，修改列名
appoint.columns = ['预约编号','预约人','soundi','facility','sproperty','refur','specification','time','num','events']
```

```python
appoint = appoint[2:].reset_index(drop=True)
appoint
```

### soundi

```python
set(appoint['soundi'])
```

```python
soundi_mapping = {
    'Excellent':2,
    'Not specified':0,
}
# 总可以选择大于等于自身隔音等级的空间
appoint['soundi']= appoint['soundi'].apply(lambda x: soundi_mapping.get(x,0)) #除上述映射其余置为0
```

### facility

```python
set(appoint['facility'])
```

```python
# 数据映射表
# facility_mapping = {
#     '0':0,
#     '1| Projector':1,
#     '1| Board':2,
#     '1| Computer':3,
#     '1| Shelves':4,
#     '1| 3D printer':5,
#     '1| Eye tracker':6,
#     '1| EEG':7,
#     '1| Audio-visual equipment':8
# }
facility_mapping = {
    '0':0,
    'Projector':1,
    '1| Board':2,
    'Computer':3,
    'Shelves':4,
    '1| 3D printer':5,
    '1| Eye tracker':6,
    '1| EEG':7,
    '1| Audio-visual equipment':8
}
# 根据映射修改列
appoint['facility'] = appoint['facility'].apply(lambda x: facility_mapping.get(x, 0)) #若存在key则映射，不能则0
```

### sproperty

```python
set(appoint['sproperty'])
```

```python
# 映射
sproperty_mapping ={
     'Close':0,
     'Open':1,
} #1open,0close,nan无要求
appoint['sproperty'] = appoint['sproperty'].apply(lambda x: sproperty_mapping.get(x,2))
# 开放可以分配到任意空间<2；没有预约要求认为预约开放空间<2
```

### refur

```python
set(appoint['refur'])
```

```python
# 映射
refur_mapping ={
     'Yes':1,
} #家具需要重新排布选择1，其余都选择0，>0或者>-1都可以选择
appoint['refur'] = appoint['refur'].apply(lambda x:refur_mapping.get(x,0))
```
